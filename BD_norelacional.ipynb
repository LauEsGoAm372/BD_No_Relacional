##**1-Leer el archivo moviedata.json**

```python
with open('/content/moviedata.json') as f:
   data = json.load(f)
```

##**2- Crear un diccionario "movies" cuya llave sea el año,  y el valor, un arreglo de diccionarios con la información de las peliculas de ese año.por ej movies['2017'] debe devolver la información de todas las películas de 2017(como un arreglo de diccionarios)**
---

```python
movies = {}
for movie in data:
    year = movie["year"]
    if year not in movies:
        movies[year] = []
    movies[year].append(movie)
```

#**3- Escribir una función def query(year)   que reciba el año y devuelva todas las películas de ese año(usando el diccionario creado en el punto anterior).**
---

```python
def query(year):
  if year in movies:
    return movies[year]
  else:
    return []
```

```python
moviesitas = query(2013)
print(moviesitas)
```

#**4- Escribir una función def scan(actor) que devuelva todas las películas donde ha actuado ese actor.Usar nuevamente el diccionario movies. Sin embaro, en este punto tendrá que realizar un ciclo for de búsqueda**
---

```python
def scan(actor):
    result = []
    for m in movies.values():
        for movie in m:
            if "actors" in movie["info"]:
              if actor in movie["info"]["actors"]:
                result.append(movie)
    return result
```

```python
actor = scan("Ben Stiller")
print(actor)
```
